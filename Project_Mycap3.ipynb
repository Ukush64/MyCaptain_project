```python
#code to perform bivariate anaylsis
import matplotlib.pyplot as plt
import numpy as np
```

```python
#define the data
productivity = np.array([1,2,4,3,2,3,3,4,5,5])
time_of_day = np.array(["8am", "10am", "12pm", "2pm", "4pm", "6pm", "8pm", "10pm", "12am","2am"])

#creating bar graph
plt.bar(time_of_day, productivity)
plt.xlabel("Time of day")
plt.ylabel("Productivity level")
plt.show()
```
